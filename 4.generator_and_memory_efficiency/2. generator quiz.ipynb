# 1. Generator

```python
def my_generator():
    while True:
        value = np.random.rand(2)
        yield (value[0], value[1])
```

```python
def generator_caller():
    while True:
        yield next(my_generator())
```

# 2. Image Generator

```python
from matplotlib import pyplot as plt
```

```python
def image_generator():
    value = np.random.rand(255, 255, 3)
    value *= 255
    print(value.shape)
    yield value
```

```python
def image_show():
    new_image = image_generator()
    yield plt.imshow(next(new_image))
```

```python
next(image_show())
```
